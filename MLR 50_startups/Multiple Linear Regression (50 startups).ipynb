```python
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_r = smf.ols("RDS~ADMS+MKTS", data = df).fit().rsquared
vif_r = 1/(1-rsq_r)

rsq_a = smf.ols("ADMS~RDS+MKTS", data = df).fit().rsquared
vif_a = 1/(1-rsq_a)

rsq_m = smf.ols("MKTS~RDS+ADMS", data = df).fit().rsquared
vif_m = 1/(1-rsq_m)
```

```python
# Putting the values in Dataframe format
d1 = {'Variables':['RDS','ADMS','MKTS'], 'Vif':[vif_r,vif_a,vif_m]}
Vif_df = pd.DataFrame(d1)
Vif_df
```

# Model Deletion Diagnostics (checking Outliers or Influencers)

```python
# Two Techniques : 1. Cook's Distance & 2. Leverage value
```

```python
# Cook's Distance: If Cook's distance > 1, then it's an outlier
(c,_) = model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot

plt.rcParams['figure.figsize'] = (20,7)
plt.stem(np.arange(len(df)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
```

```python
# Index and value of influencer where C>0.5

np.argmax(c) , np.max(c)
```

```python
# Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers

influence_plot(model);
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints

k = df.shape[1]
n = df.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
df[df.index.isin([49])] 
```

```python
# Improving the model
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))

data = df.drop(df.index[[49]], axis = 0).reset_index(drop = True)
data
```

# Model Deletion Diagnostics and Final Model

```python
model2 = smf.ols("Profit~RDS+ADMS+MKTS",data = data).fit()
```

```python
while model2.rsquared < 0.99:
    for c in [np.max(c)>1]:
        model2 = smf.ols("Profit~RDS+ADMS+MKTS", data = data).fit()
        (c,_) = model2.get_influence().cooks_distance
        c
        np.argmax(c), np.max(c)
        data = data.drop(data.index[[np.argmax(c)]], axis = 0).reset_index(drop = True)
        data
    else:
        final_model = smf.ols("Profit~RDS+ADMS+MKTS", data = data).fit()
        final_model.rsquared, final_model.aic
        print("Thus model accuracy is improved to", final_model.rsquared)
```
